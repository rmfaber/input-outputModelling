```python
# import the files and get them ready to use

wiot = pd.read_csv('wiot.csv',low_memory=False)
wiot = wiot.set_index('Unnamed: 0')
wiotDEU = pd.read_csv('wiotDEU.csv')
wiotDEU = wiotDEU[wiotDEU.columns[1:]]
iiflowmatrix = pd.read_csv('flowmatrix.csv')
iiflowmatrix = iiflowmatrix[iiflowmatrix.columns[1:]]
valueadded = pd.read_csv('valueadded.csv')
valueadded = valueadded.set_index('Unnamed: 0')
valueadded = valueadded.values
valueadded = valueadded[0]
totaloutput = wiotDEU['TOT']
expZ = pd.read_excel('expandedflowmatrix.xlsx')
```

```python
# Total intermediate goods for paper products is given by colsum.
iiflowmatrix.sum()[7]
```

```python
# calculate the inverse total output diagonal
idenmatrix = np.identity(56)
totaloutput = np.array(totaloutput)
l = []
for x in totaloutput:
    if x != 0:
        x = 1/x
    l.append(x)
invtotaloutput = np.array(l)
invx = invtotaloutput * idenmatrix
```

```python
# multiply total output diagonal x with Z (the flow matrix) to get A
A = np.matmul(iiflowmatrix,invx)
L1 = idenmatrix - A
L1DF = pd.DataFrame(L1)
# Inverse this to get the leontief 
Leontief = np.linalg.inv(L1)
# convert to pandas dataframe after the calculations, which allows for easier viewing
# and writing to excel or csv files
LeontiefDF = pd.DataFrame(Leontief)
LeontiefDF.to_excel('Leontief.xlsx')
```
